```python
import pandas as pd
import yfinance as yf
```

```python
def add_stock(ISIN='', BSEStockCode='', NSEStockCode=''):
    if ISIN != '':
        stock = yf.Ticker(ISIN)
    elif NSEStockCode != '':
        stock = yf.Ticker(NSEStockCode+'.NS')
    elif BSEStockCode != '':
        stock = yf.Ticker(BSEStockCode+'.BO')
    else:
        return
    print(stock.info)
```

```python
add_stock(NSEStockCode='VEDL')
```

# Logic to Add a new stock with all other details

```python
def create_new_stock(isin):
    try:
        stock = yf.Ticker(isin)
        print(isin)
        print(stock.info['symbol'].split('.')[0])
        print(stock.info['longName'])
    except Exception as e:
        print('Exception')
        return
```
