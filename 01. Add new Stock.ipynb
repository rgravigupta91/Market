```python
import pandas as pd
import yfinance as yf
import requests
from datetime import datetime
```

```python
def get_listing_date(ISIN, symbol):
    

    headers = {
        "User-Agent": "Mozilla/5.0"
    }
    url = f"https://www.nseindia.com/api/search/autocomplete?q={symbol}"
    response = requests.get(url, headers=headers)

    a = response.json()
    b = a['symbols'][0]['listing_date']
    ListingDate = datetime.strptime(b, "%Y-%m-%d").date()
    
    values = (ListingDate, ISIN)
    return values
```

```python
get_listing_date('INE06IR01021','OSIAHYPER')
```

```python
def add_stock(ISIN='', BSEStockCode='', NSEStockCode=''):
    if ISIN != '':
        stock = yf.Ticker(ISIN)
    elif NSEStockCode != '':
        stock = yf.Ticker(NSEStockCode+'.NS')
    elif BSEStockCode != '':
        stock = yf.Ticker(BSEStockCode+'.BO')
    else:
        return
    if "symbol" not in stock.get_info():
        raise ValueError('No Stock Exists with the given code')
    
    NSEStockCode = str(stock.info['symbol'].split('.')[0])
    print('ISIN : ', ISIN)
    print('NSE Stock Code : ', NSEStockCode)
    print('Listing date : ', get_listing_date(ISIN, NSEStockCode))
    print('Market Cap : ', stock.info['marketCap'])
    print('Outstanding Shares : ', stock.info["sharesOutstanding"])
    print('Floating Shares : ', stock.info['floatShares'])
    print('Current Price : ', stock.info['currentPrice'])
    print('Book Value : ', stock.info['bookValue'])
    print('Beta : ', stock.info['beta'])
    print('Regular Market Price : ', stock.info['regularMarketPrice'])



    return stock
```

```python
stock = add_stock(ISIN='INE205A01025', NSEStockCode='VEDL')
```
