```python
# Update Dividend for Stocks
from mymodule.DatabaseConnection import CRUD
import yfinance as yf
import pandas as pd
import numpy as np

lo_stockcode = CRUD('StockExchange_stockcode')
df_stockcode = pd.DataFrame(lo_stockcode.Read(''))

sub = lambda a: 'NS' if a == 'NSE' else 'BO'

lo_div = CRUD('dividends')
df_dividends = pd.DataFrame(lo_div.Read(''))

for i, code in df_stockcode[1:].iterrows():
    if code['StockExchange'] != 'NSE':
        continue
    query_symbol = code['Symbol'] + '.' + sub(code['StockExchange'])
    print(query_symbol)
    dat = yf.Ticker(query_symbol)
    dividend = pd.DataFrame(dat.dividends)
    if len(dat.dividends) > 0:
        dividend['ISIN'] = code['ISIN']
        dividend['RecordDate'] = dividend.index.date
        dividend['dividend'] = (dividend['Dividends'])
        df_dividends = pd.concat([df_dividends, dividend[df_dividends.columns]])

df_dividends.reset_index(drop=True, inplace=True)
lo_div.Create(df_dividends)
```

```python
import requests
from bs4 import BeautifulSoup

def get_cin_from_mca(company_name: str):
    """
    Fetches CIN (Corporate Identification Number) of an Indian company
    by scraping MCA portal search results.
    """
    try:
        # MCA company search URL
        url = "https://www.mca.gov.in/mcafoportal/viewCompanyMasterData.do"
        search_url = f"https://www.mca.gov.in/mcafoportal/findCompanyName.do"

        # Step 1: Search by name
        params = {"companyName": company_name}
        resp = requests.get(search_url, params=params, timeout=15)

        if resp.status_code != 200:
            raise Exception("Failed to reach MCA portal")

        soup = BeautifulSoup(resp.text, "html.parser")

        # Step 2: Try to find CIN in response (common pattern)
        cin_tag = soup.find(string=lambda s: "CIN" in s)  # Find line containing CIN
        if not cin_tag:
            print("CIN not found on MCA page.")
            return None

        # Step 3: Extract CIN
        cin_text = cin_tag.strip().split(":")[-1].strip()
        return cin_text

    except Exception as e:
        print(f"Error fetching CIN: {e}")
        return None

# Example
company_name = "Nestle India Limited"
cin = get_cin_from_mca(company_name)
print(f"{company_name} → CIN: {cin}")
```
