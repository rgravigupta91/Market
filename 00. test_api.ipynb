```python
# Update Dividend for Stocks
from mymodule.DatabaseConnection import CRUD
import yfinance as yf
import pandas as pd
import numpy as np

lo_stockcode = CRUD('StockExchange_stockcode')
df_stockcode = pd.DataFrame(lo_stockcode.Read(''))

sub = lambda a: 'NS' if a == 'NSE' else 'BO'

lo_div = CRUD('dividends')
df_dividends = pd.DataFrame(lo_div.Read(''))

for i, code in df_stockcode[1:].iterrows():
    if code['StockExchange'] != 'NSE':
        continue
    query_symbol = code['Symbol'] + '.' + sub(code['StockExchange'])
    print(query_symbol)
    dat = yf.Ticker(query_symbol)
    dividend = pd.DataFrame(dat.dividends)
    if len(dat.dividends) > 0:
        dividend['ISIN'] = code['ISIN']
        dividend['RecordDate'] = dividend.index.date
        dividend['dividend'] = (dividend['Dividends'])
        df_dividends = pd.concat([df_dividends, dividend[df_dividends.columns]])

df_dividends.reset_index(drop=True, inplace=True)
lo_div.Create(df_dividends)
```

```python
# To Upate Daily stock Price
from mymodule.DatabaseConnection import CRUD
import yfinance as yf

lo_stockcode = CRUD('StockExchange_stockcode')
df_stockcode = pd.DataFrame(lo_stockcode.Read(''))

lo_daily_update = CRUD("StockDailyUpdate")

sub = lambda a: 'NS' if a == 'NSE' else 'BO'

def get_exchange_from_ticker(ticker: yfinance.ticker.Ticker) -> str:
    if str(ticker.info['symbol']).endswith(".NS"):
        return "NSE"
    elif str(ticker.info['symbol']).endswith(".BO"):
        return "BSE"
    return "UNKNOWN"


for i, code in df_stockcode.iterrows():
    query_symbol = code['Symbol'] + '.' + sub(code['StockExchange'])
    dat = yf.Ticker(query_symbol)
    cedl = dat.history(period='10y')
    #print(query_symbol)
    if cedl.shape[0] > 0:
        cedl["Date"] = cedl.index
        cedl["Date"] = cedl["Date"].dt.date
        cedl.insert(0, "Date", cedl.pop("Date"))
        #cedl.pop("Dividends")
        #cedl.pop("Stock Splits")
        cedl["delivery"] = 0
        cedl.insert(0, "Symbol",dat.info["symbol"].split(".")[0])
        cedl.insert(0, "StockExchange",get_exchange_from_ticker(dat))
        cedl = cedl[['StockExchange','Symbol','Date','Open','High','Low','Close','Volume','delivery']]
        cedl.fillna(0, inplace=True)
        lo_daily_update.Create(cedl)
        print('Row ',i,' : ',query_symbol, " : ",cedl.shape[0], " records inserted")
        #del cedl
```
